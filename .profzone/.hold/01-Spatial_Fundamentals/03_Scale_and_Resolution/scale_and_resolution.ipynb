
# Scale & Resolution

As soon as we start visualizing or analyzing spatial data, two words show up everywhere:

**scale** and **resolution**.

They sound similar.  
They are not the same.

Confusing them leads to bad maps, bad analysis, and bad conclusions.



## What Is Scale?

Scale describes **how much of the world we are looking at**.

- Large scale → small area, lots of detail  
- Small scale → large area, less detail  

Yes, this feels backwards at first.  
That’s normal.



## What Is Resolution?

Resolution describes **how detailed the data itself is**.

Resolution answers questions like:
- How small is the smallest feature I can represent?
- How fine is the grid or sampling?


```python

# Imagine a simple 1D "world" sampled at different resolutions

world = list(range(0, 100))

# Low resolution: sample every 10 units
low_res = world[::10]

# High resolution: sample every unit
high_res = world[::1]

low_res, high_res[:10]
```


## Same World, Different Resolution

Both datasets represent the *same* underlying space.
They simply describe it at different levels of detail.


```python

len(low_res), len(high_res)
```


## Why Resolution Matters

Higher resolution:
- Uses more storage
- Takes longer to process
- Is not always better

Lower resolution:
- Is faster
- Is simpler
- May hide important patterns



## Scale vs Resolution (Together)

These two concepts interact, but they are independent:

- You can view **small-scale** data at **high resolution**
- You can view **large-scale** data at **low resolution**


```python

# Simple example: aggregating data to change resolution

def aggregate(data, step):
    return data[::step]

aggregate(world, 5)[:10]
```


## Common Misconceptions

- More resolution always means better analysis ❌
- Zooming in creates new information ❌
- Scale and resolution mean the same thing ❌
